## Criando a coluna duplicatas

```python
df = df_data.groupby('lesion_id').count()
```

```python
df.head()
```

```python
df = df[df['image_id'] == 1] # Imagens somente com uma associação
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
def identificando_duplicatas(x):
    lista_unica = list(df['lesion_id'])
    if x in lista_unica:
        return 'Nao'
    else:
        return 'Sim'
```

```python
df_data['Tem_Duplicatas'] = df_data['lesion_id']
df_data['Tem_Duplicatas'] = df_data['Tem_Duplicatas'].apply(identificando_duplicatas)

df_data.head()
```

## Adicionando a extensão no nome da imagem

```python
def append_ext(fn):
    return fn+".jpg"
```

```python
df_data['image_id']=df_data['image_id'].apply(append_ext)
```

# Separando os folds manualmente

```python
df_val = df_data[df_data['Tem_Duplicatas'] == 'Nao']
```

```python
df_val.info()
```

```python
df_val.head()
```

## Criando a divisão

```python
import sklearn
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits = 5, random_state = 43, shuffle = True) 
Y = df_val['dx']
n = len(Y)
```

```python
df_val.head()
```

## Criando as colunas para cada fold

```python
df_data['fold_1'] = df_data['image_id']
```

```python
df_data.head()
```

```python
df_data['fold_2'] = df_data['image_id']
df_data['fold_3'] = df_data['image_id']
df_data['fold_4'] = df_data['image_id']
df_data['fold_5'] = df_data['image_id']
```

```python
df_data.head()
```

```python
df.info()
```

## Realizando a criação dos folds, nesse caso 5

```python
fold_n = 1
for train_index, val_index in skf.split(np.zeros(n), Y):
    
    df_validador = df_val.iloc[val_index]
    
    def identificando_vals(x):
        
        val_list = list(df_validador['image_id'])
        
        if str(x) in val_list:
            return 'val'
        else:
            return 1
        
    df_data['fold_'+str(fold_n)] = df_data['fold_'+str(fold_n)].apply(identificando_vals)
    fold_n += 1 
```

```python
df_data.fold_1 = df_data.fold_1.map({1: 'train', 'val':'val'})
```

```python
df_data.head()
```

```python
df_data.fold_2 = df_data.fold_2.map({1: 'train', 'val':'val'})
df_data.fold_3 = df_data.fold_3.map({1: 'train', 'val':'val'})
df_data.fold_4 = df_data.fold_4.map({1: 'train', 'val':'val'})
df_data.fold_5 = df_data.fold_5.map({1: 'train', 'val':'val'})
```

# Separando os dados em folders

```python
def criando_diretorios (num):
    base_dir = 'C:\\Users\\doxda\\Downloads\\Dados-TCC-Melo\\Dados\\fold_'+str(num)
    os.mkdir(base_dir)

    # train_dir
    train_dir = os.path.join(base_dir, 'train_dir')
    os.mkdir(train_dir)

    # val_dir
    val_dir = os.path.join(base_dir, 'val_dir')
    os.mkdir(val_dir)
    
        # create new folders inside train_dir
    nv = os.path.join(train_dir, 'nv')
    os.mkdir(nv)
    mel = os.path.join(train_dir, 'mel')
    os.mkdir(mel)
    bkl = os.path.join(train_dir, 'bkl')
    os.mkdir(bkl)
    bcc = os.path.join(train_dir, 'bcc')
    os.mkdir(bcc)
    akiec = os.path.join(train_dir, 'akiec')
    os.mkdir(akiec)
    vasc = os.path.join(train_dir, 'vasc')
    os.mkdir(vasc)
    df = os.path.join(train_dir, 'df')
    os.mkdir(df)


    # create new folders inside val_dir
    nv = os.path.join(val_dir, 'nv')
    os.mkdir(nv)
    mel = os.path.join(val_dir, 'mel')
    os.mkdir(mel)
    bkl = os.path.join(val_dir, 'bkl')
    os.mkdir(bkl)
    bcc = os.path.join(val_dir, 'bcc')
    os.mkdir(bcc)
    akiec = os.path.join(val_dir, 'akiec')
    os.mkdir(akiec)
    vasc = os.path.join(val_dir, 'vasc')
    os.mkdir(vasc)
    df = os.path.join(val_dir, 'df')
    os.mkdir(df)
    
    return train_dir, val_dir
```

```python
import itertools
import shutil

for fold_n in range(1,6):
    train_dir, val_dir = criando_diretorios (fold_n)
    df_train = df_data[df_data['fold_'+str(fold_n)] == 'train']
    df_val = df_data[df_data['fold_'+str(fold_n)] == 'val']

    train_list = list(df_train.index)
    val_list = list(df_val.index)
    
    for image in train_list:
        
        label = df_data.loc[image, 'dx']
        src = os.path.join('C:\\Users\\doxda\\Downloads\\Dados-TCC-Melo\\Dados\\ISIC2018_Task3_Training_Input_Modified', image)
        dst = os.path.join(train_dir, label, image)
        
        shutil.copyfile(src, dst)
        
    for image in val_list:
        
        label = df_data.loc[image, 'dx']
        src = os.path.join('C:\\Users\\doxda\\Downloads\\Dados-TCC-Melo\\Dados\\ISIC2018_Task3_Training_Input_Modified', image)
        dst = os.path.join(val_dir, label, image)
        
        shutil.copyfile(src, dst)
```
