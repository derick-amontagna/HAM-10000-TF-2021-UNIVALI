```python
classes = list(df.Targets.unique())
```

```python
data = _datagen.flow_from_dataframe(dataframe=df,
                                    directory=r'C:\Users\doxda\Downloads\Dados-TCC-Melo\ISIC2018_Task3_Training_Input',
                                    x_col='image',
                                    y_col='Targets',
                                    classes=classes,
                                    batch_size=1,
                                    seed=42,
                                    shuffle=True,
                                    target_size=(224,224),
                                    class_mode='categorical')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,15))
for i in range(2):
    for j in range(2):
        image, label = data.next()
        ax[i, j].imshow(image[0].astype('uint8'))
        ax[i, j].axis('off')
```

# Modo 1

```python
path = r'C:\Users\doxda\Downloads\Dados-TCC-Melo\ISIC2018_Task3_Training_Input_Modified\ISIC_0024306.jpg'
img = tf.keras.preprocessing.image.load_img(path, color_mode="rgb", target_size=(224,224))
img = tf.keras.preprocessing.image.img_to_array(img)
img.shape
def augment (image):
    image = tf.image.random_saturation(image, 1, 3)
    image = image/255
    return image
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,15))
mod = augment(img)
ax[0].set_title('(a) Original',fontdict = {'fontsize': 30})
ax[0].imshow(img.astype('uint8'))
ax[0].axis('off')
ax[1].set_title('(b) Modificada',fontdict = {'fontsize': 30})
ax[1].imshow(mod)
ax[1].axis('off')
fig.tight_layout()
plt.show()
```

```python
fig.savefig('f1.jpg', dpi=1200, bbox_inches='tight')
```

```python
path = r'C:\Users\doxda\Downloads\Dados-TCC-Melo\ISIC2018_Task3_Training_Input_Modified\ISIC_0024306.jpg'
img = tf.keras.preprocessing.image.load_img(path, color_mode="rgb", target_size=(224,224))
img = tf.keras.preprocessing.image.img_to_array(img)
img = np.expand_dims(img, axis=0)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,15))
_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255, horizontal_flip=True, vertical_flip=True)
data = _datagen.flow(img, batch_size=1)
mod = next(iter(data))
ax[0].set_title('(a) Original',fontdict = {'fontsize': 30})
ax[0].imshow(img[0].astype('uint8'))
ax[0].axis('off')
ax[1].set_title('(b) Modificada',fontdict = {'fontsize': 30})
ax[1].imshow(mod[0])
ax[1].axis('off')
fig.tight_layout()
plt.show()
```

# Random CRop

```python
path = r'C:\Users\doxda\Downloads\Dados-TCC-Melo\ISIC2018_Task3_Training_Input_Modified\ISIC_0024306.jpg'
img = tf.keras.preprocessing.image.load_img(path, color_mode="rgb")
img = tf.keras.preprocessing.image.img_to_array(img)
img = np.expand_dims(img, axis=0)
img.shape
```

```python
def augment (image):
    image = tf.image.random_crop(image, size = [224, 224, 3])
    return image
_datagen = tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=augment)
data = _datagen.flow(img, batch_size=1)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(5,5))
for i in range(2):
    for j in range(2):
        image = next(iter(data))
        ax[i, j].imshow(image[0].astype('uint8'))
        ax[i, j].axis('off')
fig.tight_layout()
```

```python
fig.savefig('f1.jpg', dpi=1200, bbox_inches='tight')
```
