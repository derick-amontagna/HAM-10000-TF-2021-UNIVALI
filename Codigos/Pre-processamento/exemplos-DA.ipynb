```python
import numpy as np
import pandas as pd
import os

import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

from tensorflow.keras import layers
import PIL
from matplotlib import image
from matplotlib import pyplot
```

```python
image_path = r'C:\Users\doxda\Downloads\Dados-TCC-Melo\Pre-processamento\mel_16.jpg'
```

```python
image=tf.io.read_file(image_path)
image = tf.io.decode_jpeg(image)
```

```python
def visualize(original, augmented ,aug2):
  fig = plt.figure()
  plt.subplot(1,3,1)
  plt.title('Original')
  plt.axis('off')
  plt.imshow(original)

  plt.subplot(1,3,2)
  plt.title('Virada Horizontal')
  plt.axis('off')
  plt.imshow(augmented)

  plt.subplot(1,3,3)
  plt.title('Virada Vertical')
  plt.axis('off')
  plt.imshow(aug2)
 

image =tf.io.read_file(image_path)
image = tf.io.decode_jpeg(image)

flipped = tf.image.flip_left_right(image)
flipped2 = tf.image.flip_up_down(image)
visualize(image, flipped, flipped2)
plt.savefig('virada.png')
```

```python
def visualize(original, augmented):
  fig = plt.figure()
  plt.subplot(1,2,1)
  plt.title('Original')
  plt.axis('off')
  plt.imshow(original)

  plt.subplot(1,2,2)
  plt.title('Contraste aleatria')
  plt.axis('off')
  plt.imshow(augmented)

image=tf.io.read_file(image_path)
image = tf.io.decode_jpeg(image)

flipped = tf.image.random_contrast(image, 0.8, 1.2)
visualize(image, flipped)
plt.savefig('consta.png')
```
